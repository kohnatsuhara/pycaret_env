```python
import pandas as pd
from pycaret.regression import *
```

```python
import pycaret
pycaret.__version__
```

```python
from pycaret.datasets import get_data 
boston_data = get_data('boston')
```

```python
exp = setup(boston_data, target = 'medv', ignore_features = None)
```

```python
best = exp.compare_models()
```

```python
exp.plot_model(best, plot = 'residuals')
```

```python
exp.plot_model(best, plot = 'error')
```

```python
exp.plot_model(best, plot = 'feature')
```

```python
dashboard(
    best
    ,display_format='dash' # dashモジュールで起動。ウェブサーバーが立ち上がるのでブラウザでlocalhost:8050にアクセス
    ,run_kwargs={"host":"0.0.0.0"} # pycaret内部でExplainerDashboard.run()が実行されており、runに渡す引数。dashboardを立ち上げるIPを指定
)
# 実行後、ブラウザにて「localhost:8050」にアクセス
```

```python
best.predict(boston_data)# 予測する場合predict()にデータを入れて実行
```

```python
import pandas as pd
from pycaret.regression import *
```

```python
import pycaret
pycaret.__version__
```

```python
from pycaret.datasets import get_data 
boston_data = get_data('boston')
```

```python
exp = setup(boston_data, target = 'medv', ignore_features = None)
```

```python
best = exp.compare_models()
```

```python
exp.plot_model(best, plot = 'residuals')
```

```python
exp.plot_model(best, plot = 'error')
```

```python
exp.plot_model(best, plot = 'feature')
```

```python
dashboard(
    best
    ,display_format='dash' # dashlocalhost:8050
    ,run_kwargs={"host":"0.0.0.0"} # pycaretExplainerDashboard.run()rundashboardIP
)
# localhost:8050
```

```python
best.predict(boston_data)# predict()
```
